```python
wnp_metadata = pd.read_csv('subset_cyclones/wnp_era5_metadata.txt', sep = '\t')
```

```python
wnp_metadata.shape[0]
```

```python
wnp_metadata
```

```python
for i in range(0,wnp_metadata.shape[0]):
    
    domain = [wnp_metadata['domain_latMax'][i],
        wnp_metadata['domain_lonMin'][i],
        wnp_metadata['domain_latMin'][i],
        wnp_metadata['domain_lonMax'][i]]
    
    c.retrieve('reanalysis-era5-pressure-levels', {
        'variable'      : 'geopotential',
        'pressure_level': '500',
        'product_type'  : 'reanalysis',
        'date'          : wnp_metadata['date'][i],
        'time'          : wnp_metadata['time'][i],
        'area'          : list(map(float, domain)),
        'format'        : 'netcdf'
    }, './ERA5_data/wnp/wnp_original/{wnp_id}.nc'.format(wnp_id = wnp_metadata['wnp_id'][i]))
```
